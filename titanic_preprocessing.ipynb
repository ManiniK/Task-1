```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('Titanic-Dataset[1].csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True) #using one-hot encoding
```

```python
scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
#Visualizing using boxplots
sns.boxplot(data=df[['Age', 'Fare']])
plt.show()
```

```python
#Removing outliers using IQR method
Q1 = df[['Age', 'Fare']].quantile(0.25)
Q3 = df[['Age', 'Fare']].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[['Age', 'Fare']] < (Q1 - 1.5 * IQR)) |
          (df[['Age', 'Fare']] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Shape after removing outliers
print("Shape after removing outliers:", df.shape)
```

```python
#Re-plotting box-plot
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("Boxplot After Removing Outliers")
plt.show()
```

```python
df.to_csv('titanic_cleaned.csv', index=False)
```
